# TLB-8800 discovery

Run this notebook from the **Laser** folder (`hardware/Laser`).

- `discovery.py` + `supported_models.json` — find lasers on COM ports
- `newfocus/TLB8800.py` — connect and control the laser

```python
import sys
from pathlib import Path

root = Path.cwd()
if str(root) not in sys.path:
    sys.path.insert(0, str(root))

from discovery import LaserDiscovery, list_usb_ports
from newfocus import program_data_lab_root

discovery = LaserDiscovery()
print("Config:", discovery.config_path)
print("Supported:", [m.model for m in discovery.supported_models])
print("Log root:", program_data_lab_root())
print("USB COM ports:", list_usb_ports())
lasers = discovery.discover()
lasers
```

```python
from newfocus import TLB8800

port, laser_id = next(iter(lasers.items()))
print("Port:", port)
print("ID:", laser_id)

laser = TLB8800.connect(port)
print("Log file:", laser.log_path)

specs = laser.specs
print("Current control:", specs.current_control)
print("Wavelength range:", specs.wavelength_min, "-", specs.wavelength_max, "nm")
print("Scan speed range:", specs.scan_speed_min, "-", specs.scan_speed_max)
print("Power setpoint:", specs.power)
print("Current:", specs.current, "mA (max", specs.current_max, ")")
laser
```

```python
print("OPC:", laser.read.operation_complete())
print("Interlock:", laser.read.interlock_state())
print("Laser out:", laser.read.laser_output())
```

```python
print(laser.ON())
print("Laser out:", laser.read.laser_output())
```
